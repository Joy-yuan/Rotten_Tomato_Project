```python
import requests
from bs4 import BeautifulSoup
import pprint as p
import pandas as pd
import numpy as np
from selenium import webdriver
import time
from selenium.webdriver.common.by import By
```

```python
big_list=[]

    
url = "https://www.imdb.com/list/ls000632442/?sort=list_order,asc&mode=grid&page=1&ref_=nmls_vw_grd"
html_data = requests.get(url)
if html_data.status_code == 200:
    python_data = BeautifulSoup(html_data.content,'lxml')
director_tag=python_data.find_all('div',class_="lister-item mode-grid")
    
for i in director_tag:
    small_list=[1,2]
    dir_rank=i.find('div',class_="title").find('span').get_text()[:-1]
    dir_name=i.find('div',class_="title").find('a').get_text()
    small_list[0]=dir_rank[:-1].strip(',')
    small_list[1]=dir_name.strip('\n')
    big_list.append(small_list)
big_list
```

```python
df = pd.DataFrame(big_list)
df.columns=['rank','director']
df['director_score']=df['rank'].copy()
df['director_score']=df['director_score'].apply(lambda x: 10-int(x)//6)
df
```

```python
df.to_csv('director_score.csv')
```
