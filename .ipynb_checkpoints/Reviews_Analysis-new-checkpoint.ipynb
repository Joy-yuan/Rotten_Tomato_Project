### Web Scrapping and Exporting Data

```python
def get_review(link):               #returns all top critic reviews for one movie
    link = link
    response = requests.get(link)
    review_collection = []
    if response.status_code == 200:
        try:
            result_page = BeautifulSoup(response.content,'lxml')
            all_review = result_page.find('div',class_='review_table').find_all('div',class_='the_review')
            for review in all_review:
                review_collection.append(review.get_text())
        except:
            return get_review(link)
    return review_collection
```

```python
def get_movie_review(link,year,name):       #returns reviews of all top movies for one year
    
    response1 = requests.get(link)
    if response1.status_code == 200:
        result_page1 = BeautifulSoup(response1.content,'lxml')
        body = result_page1.find('section', id="contentReviews")
        try:
            sibody = body.find('a',class_='small unstyled subtle articleLink')
            siibody = sibody.next_sibling.next_sibling.get('href')
            link = 'https://www.rottentomatoes.com' + siibody

        except:
            return get_movie_review(link,year,name)
            
            
        
        review_collection = get_review(link)
        review = [name, year, review_collection]

    return review

```

```python
all_reviews = []
for year in range(2014,2019):                     #getting 5 years of reviews
    page_link = 'https://www.rottentomatoes.com/top/bestofrt/?year='+str(year)
    init_response = requests.get(page_link)
    if init_response.status_code == 200:
        topmovie_page = BeautifulSoup(init_response.content,'lxml')
        movie_link_list= topmovie_page.find('table',class_='table').find_all('a',class_="unstyled articleLink" )
        for movie in movie_link_list:
            movie_link = 'http://www.rottentomatoes.com' + movie.get('href')
            movie_name = movie.get_text()[13:-7]      
            #print(movie_name)
            #time.sleep(2)
            review = get_movie_review(movie_link,year,movie_name)
            all_reviews.append(review)
        
```

### Preparation for text mining

#### 1. NRC sentiment data
http://sentiment.nrc.ca/lexicons-for-research/NRC-Emotion-Lexicon.zip

```python
def get_nrc_data():
    nrc = "NRC-Emotion-Lexicon-Wordlevel-v0.92.txt"
    count=0
    emotion_dict=dict()
    with open(nrc,'r') as f:
        all_lines = list()
        for line in f:
            if count < 46:
                count+=1
                continue
            line = line.strip().split('\t')
            if int(line[2]) == 1:
                if emotion_dict.get(line[0]):
                    emotion_dict[line[0]].append(line[1])
                else:
                    emotion_dict[line[0]] = [line[1]]
    return emotion_dict
emotion_dict = get_nrc_data()
```

### Sentiment Analysis

```python
def emotion_analyzer(text,emotion_dict=emotion_dict):
    #Set up the result dictionary
    emotions = {x for y in emotion_dict.values() for x in y}  
    emotion_count = dict()
    for emotion in emotions:
        emotion_count[emotion] = 0

        total_words = len(text.split())     
    for word in text.split():     #for each word
        if emotion_dict.get(word):     #if it's in the dict database
            for emotion in emotion_dict.get(word):      #then for all emotions attached to this word
                emotion_count[emotion] += 1/len(text.split())        #add the percentage to each emotion count
    return emotion_count       #return a dict with percentage of each emotion for this text
```

```python
emotion_analyzer(data['Reviews'][0])
```

```python
output = pd.DataFrame(columns=['Reviews','Fear','Trust','Negative',
                           'Positive','Joy','Disgust','Anticipation',
                           'Sadness','Surprise'],)

for i in range(0,len(data['Reviews'])):
    result = emotion_analyzer(data['Reviews'][i])
    
    output.loc[data['Movie_Name']] = [result['fear'],result['trust'],
                  result['negative'],result['positive'],result['joy'],result['disgust'],
                  result['anticipation'],result['sadness'],result['surprise']]
```

```python
def comparative_emotion_analyzer(df,object_name="Reviews",print_output=False):
    if print_output:
        print("%-20s %1s\t%1s %1s %1s %1s   %1s %1s %1s %1s"%(object_name,
                                                              "fear","trust","negative","positive",
                                                              "joy","disgust","anticip", "sadness",
                                                              "surprise"))       #printing title
    import pandas as pd
    df1 = pd.DataFrame(columns=[object_name,'Fear','Trust','Negative',
                           'Positive','Joy','Disgust','Anticipation',
                           'Sadness','Surprise'],)
    df.set_index(object_name,inplace=True)
    
    output = df1
    
    for i,review in df['Reviews'].items():
        result = emotion_analyzer(review)    #analyze emotions
        if print_output:
            print("%-20s %1.2f\t%1.2f\t%1.2f\t%1.2f\t%1.2f\t%1.2f\t%1.2f\t%1.2f\t%1.2f"%(
                df[1][0:20],result['fear'],result['trust'],
                  result['negative'],result['positive'],result['joy'],result['disgust'],
                  result['anticipation'],result['sadness'],result['surprise']))
        df.loc[df['Movie_Name']] = [result['fear'],result['trust'],
                  result['negative'],result['positive'],result['joy'],result['disgust'],
                  result['anticipation'],result['sadness'],result['surprise']]
    return output               
#And test it        
comparative_emotion_analyzer(data,print_output=True)
```

### Text Summarization

```python
summary_collection = []
for i in range(0,len(data['Reviews'])):
    text = data['Reviews'][i]
    temp_str = "%r"%text
    #print(temp_str)
    striptext = temp_str.replace('",','')
    striptext = striptext.replace("','",' ')
    striptext = striptext.replace('...','')
    striptext = striptext.replace('[','')
    striptext = striptext.replace(']','')
    striptext = striptext.replace("\\",'')
    striptext = striptext.replace("\'",'')
    striptext = striptext.replace("\',",'')
    striptext = striptext.replace(', "',' ')
    striptext = striptext.replace(".,",'.')
    striptext = striptext.replace("\\\\", '')
    summary = gensim.summarization.summarize(striptext, word_count=100) 
    summary_collection.append(summary)
summary_collection = pd.Series(summary_collection)
```

```python
new_data = pd.concat([data,summary_collection],axis=1)
new_data.columns = ['Movie_Name','Year','Reviews','Summary']
```
